# Applications of Cartesian Products

```python
import pandas as pd
```

## 1. Redo the excercises above but this time, show the Cartesian Product for data_2 $x$ data_1.
---

```python
# Generate dataset A
data_1 = pd.DataFrame({'A':[3,4],'B':[5,6]})
data_1
```

```python
# Generate dataset B
data_2 = pd.DataFrame({'C':[7,8]})
data_2
```

```python
#Create Cartesian Product
data_3 = data_2.merge(data_1, how='cross')
data_3
```

## 2. Reading and using CSV Files

```python
# Reading CSV Files
longData_1 = pd.read_csv("electronic-card-transactions-july-2023.csv")
data_1 = longData_1.head(50)

longData_2 = pd.read_csv("wholesale-trade-survey-march-2023-quarter.csv")
data_2 = longData_2.head(50)
```
